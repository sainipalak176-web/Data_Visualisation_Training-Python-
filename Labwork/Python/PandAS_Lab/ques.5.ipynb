```python
import pandas as pd

# Dataset
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# Add both Series
result = s1 + s2
print("After Addition:")
print(result)

# Replace NaN with 0 and recompute
result_fixed = s1.add(s2, fill_value=0)
print("\nAfter Replacing NaN with 0:")
print(result_fixed)
```
