```python
import pandas as pd

# Dataset
names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']
s = pd.Series(names)

# Convert all names to uppercase
upper_names = s.str.upper()
print("Uppercase Names:")
print(upper_names)

# Find names containing letter 'a'
contains_a = s[s.str.contains('a')]
print("\nNames containing 'a':")
print(contains_a)

# Replace 'anuj' with 'Anuj Kumar'
replaced = s.replace('anuj', 'Anuj Kumar')
print("\nAfter Replacement:")
print(replaced)
```
