```python
import pandas as pd

# Dataset
sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

s = pd.Series(sales, index=days)

# Best and worst sales day
best_day = s.idxmax()
worst_day = s.idxmin()

print("Best Sales Day:", best_day, "-", s[best_day])
print("Worst Sales Day:", worst_day, "-", s[worst_day])

# Total and average sales
total_sales = s.sum()
avg_sales = s.mean()

print("\nTotal Sales:", total_sales)
print("Average Sales:", avg_sales)

# Apply 20% bonus where sales > 400
bonus_sales = s.apply(lambda x: x * 1.2 if x > 400 else x)

print("\nSales after 20% bonus (if >400):")
print(bonus_sales)

# Sort sales in descending order
sorted_sales = s.sort_values(ascending=False)

print("\nSales in Descending Order:")
print(sorted_sales)
```
