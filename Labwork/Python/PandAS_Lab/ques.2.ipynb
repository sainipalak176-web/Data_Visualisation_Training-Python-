```python
import pandas as pd

salary = [25000, 32000, 28000, 40000, 50000]
emp = ['A', 'B', 'C', 'D', 'E']

s = pd.Series(salary, index=emp)

# Apply conditions safely
s = s.apply(lambda x: x*1.10 if x < 30000 else (x*0.95 if x > 45000 else x))

# Total salary
total_salary = s.sum()

print("Updated Salary:\n", s)
print("\nTotal Salary Expense:", total_salary)
```
