```python
import pandas as pd

# Dataset
data = [5, None, 15, 25, None, 35, 45]
s = pd.Series(data)

# Chained Operations
result = (
    s.fillna(s.median())      # 1. Fill missing with median
     .loc[lambda x: x > 20]   # 2. Filter values > 20
     .sort_values(ascending=False)  # 3. Sort descending
     .reset_index(drop=True)  # 4. Reset index
)

print(result)
```
