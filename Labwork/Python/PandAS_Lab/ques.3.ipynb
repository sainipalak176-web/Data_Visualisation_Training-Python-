```python
import pandas as pd

# Dataset
data = [10, None, 30, None, 50, 60, None]
s = pd.Series(data)

# 1. Count missing values before filling
print("Missing values before filling:", s.isna().sum())

# 2. Fill missing values with mean
mean_value = s.mean()
s_filled = s.fillna(mean_value)

print("\nSeries after filling missing values:")
print(s_filled)

# 3. Count missing values after filling
print("\nMissing values after filling:", s_filled.isna().sum())

# 4. Drop remaining nulls (if any)
s_clean = s_filled.dropna()

print("\nFinal cleaned Series:")
print(s_clean)
```
