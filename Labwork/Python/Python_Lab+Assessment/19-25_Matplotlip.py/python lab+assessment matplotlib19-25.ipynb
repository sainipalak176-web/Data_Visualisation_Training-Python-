```python
import matplotlib.pyplot as plt

# Input
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

# Output
plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input
products = ['A', 'B', 'C', 'D', 'E']
sales = [10, 25, 15, 30, 20]

# Output
plt.bar(products, sales)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Bar Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 25]

# Output
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
# Output: Pie chart showing percentage distribution of categories
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Input
data = np.random.randint(1, 50, 10)

# Output
plt.hist(data)
plt.title("Histogram")
plt.show()
# Output: Histogram showing distribution of random data
```

```python
import matplotlib.pyplot as plt

# Input
x = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]
y2 = [1, 3, 5, 7]

# Output
plt.plot(x, y1, label="Dataset 1")
plt.plot(x, y2, label="Dataset 2")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Combined Plot")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input
x = [1, 2, 3, 4, 5]
y = [10, 25, 15, 30, 20]

# Output
plt.scatter(x, y)

# Highlight max point
max_y = max(y)
max_x = x[y.index(max_y)]
plt.scatter(max_x, max_y, color='red')

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter Plot with Max Point")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Input
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [100, 150, 130, 170, 160]

# Output
plt.plot(months, revenue)
plt.title("Monthly Revenue")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
# Output: Line plot with title, axis labels, and grid
```
